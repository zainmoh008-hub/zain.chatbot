```python
df = pd.read_csv("/content/أسئلة_عامة_ومنطقية_150_سؤال (1).csv")#قراة الملف و عدد الاسئة فيه

questions = df["السؤال"].tolist()
answers = df["الإجابة المفصلة"].tolist()

len(df)
```

```python
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")#نموذج الفهم الذكي عشان يخلي الروبوت يفهم جميع الاسئلة حتى لو ما كانت حرفي يعني حتى لو قريب منها
```

```python
question_embeddings = model.encode(questions, convert_to_tensor=True)# تحويل الاسئلة من نص الى تمثيل عددي حتى يستطيع النموذج فهم المعنى ومقارنة تشابه الاسئلة
```

```python
def chatbot(user_question):# دالة تستقبل سؤال الطالب وتحسب اقرب سؤال في قاعدة البيانات بناء على التشابه ثم ترجع الجواب المناسب
    user_embedding = model.encode(user_question, convert_to_tensor=True)
    similarities = util.cos_sim(user_embedding, question_embeddings)
    best_match_idx = torch.argmax(similarities)
    return answers[best_match_idx]
```
